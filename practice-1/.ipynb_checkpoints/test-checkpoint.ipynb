```python
import os
import shutil
import pathlib 


def tree(directory):
    print(f'-- {directory}')
    for path in sorted(directory.rglob('*')):
        depth = len(path.relative_to(directory).parts)
        spacer = '    ' * depth
        print(f'{spacer}-- {path.name}')        
def rules():
    print(" üòÅWellcomeüòÅ\nRules this program:")
    print(' ‚úìstart - –¥–ª—è –Ω–∞—á–∞–ª–∞ —Ä–∞–±–æ—Ç—ã // exit - –¥–ª—è –≤—ã—Ö–æ–¥–∞')
    print(' ‚úì–ü–µ—Ä–µ–º–µ—â–µ–Ω–∏–µ –ø–æ –∫–∞—Ç–∞–ª–æ–≥—É —Å –ø–æ–º–æ—â—å—é "." - –≤–Ω–∏–∑, ".." - –≤–≤–µ—Ä—Ö ')
    print(' ‚úì–î–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –∫–∞—Ç–∞–ª–æ–≥–∞ - 1\n',
          '‚úì–î–ª—è –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∫–∏ –∫–∞—Ç–∞–ª–æ–≥–∞/—Ñ–∞–π–ª–∞ - 2\n',
          '‚úì–î–ª—è —É–¥–∞–ª–µ–Ω–∏—è –∫–∞—Ç–∞–ª–æ–≥–∞ - 3\n',
          '‚úì–î–ª—è —Å–æ–∑–¥–∞–Ω–∏—è —Ñ–∞–π–ª–∞ - 4\n',
          '‚úì–î–ª—è –∑–∞–ø–∏—Å–∏ –≤ —Ñ–∞–π–ª - 5\n',
          '‚úì–î–ª—è –ø—Ä–æ—á—Ç–µ–Ω–∏—è —Ñ–∞–π–ª–∞ -  6\n',
          '‚úì–î–ª—è —É–¥–∞–ª–µ–Ω–∏—è —Ñ–∞–π–ª–∞ - 7\n',
          '‚úì–î–ª—è –∫–æ–ø–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ–∞–π–ª–∞ - 8\n',
          '‚úì–î–ª—è –ø–µ—Ä–µ–º–µ—â–µ–Ω–∏—è —Ñ–∞–π–ª–∞ - 9\n'
         )  
def creation():
    byl = True
    cwd = os.getcwd()
    st = str(input('–í–≤–µ–¥–∏—Ç–µ start –¥–ª—è –Ω–∞—á–∞–ª–∞: '))
    if st == 'start':
        while byl:
            a = input("–í—ã–±–µ—Ä–µ—Ç–∏ –¥–µ–π—Å—Ç–≤–∏–µ:")
            print(a)
            if a=='1':
                try:
                    os.mkdir(str(input(":")))
                    print('–ö–∞—Ç–∞–ª–æ–≥ —É—Å–ø–µ—à–Ω–æ –±—ã–ª —Å–æ–∑–¥–∞–Ω')
                    print(tree(pathlib.Path.cwd()))
                except Exception as e:
                    print(e)
            elif a =='2':
                q=str(input(':'))
                w=str(input(':'))
                os.rename(q, w)
                print('–ö–∞—Ç–∞–ª–æ–≥/—Ñ–∞–π–ª —É—Å–ø–µ—à–Ω–æ –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω')
                print(tree(pathlib.Path.cwd()))
            elif a=='3':
                try:
                    b = shutil.rmtree(str(input("–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ –∫–∞—Ç–æ–ª–∞–≥–∞:")))
                    print('–ö–∞—Ç–∞–ª–æ–≥ —É—Å–ø–µ—à–Ω–æ —É–¥–∞–ª–µ–Ω')
                    print(tree(pathlib.Path.cwd()))
                except Exception as e:
                    print(e)                    
            elif a=='..' and cwd == os.getcwd():
                print('–ù–∞–∑–∞–¥ –Ω–µ–ª—å–∑—è, —ç—Ç–æ –≥–ª–∞–≤–Ω–∞—è –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏—è')
            elif a=='..' and cwd !=os.getcwd():
                os.chdir('..')
                print(tree(pathlib.Path.cwd()))
            elif a=='.':
                try:
                    d = input('–í–≤–µ–¥–∏—Ç–µ –ø—É—Ç—å –∫ –∫–∞—Ç–∞–ª–æ–≥—É:')
                    os.chdir(str(d))
                    print(tree(pathlib.Path.cwd()))
                except Exception as e:
                    print(e)
            elif a=='4':
                my_file=str(input('–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞: '))
                my_file+='.txt'
                file_res=open(my_file,'w')
                file_res.close()
                print('–§–∞–π–ª —É—Å–ø–µ—à–Ω–æ —Å–æ–∑–¥–∞–Ω')
                print(tree(pathlib.Path.cwd()))
            elif a=='5':
                try:
                    my_file=str(input('–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞: '))
                    my_file+='.txt'
                    file_res=open(my_file,'w+')
                    file_res.write(str(input('–í–≤–µ–¥–∏—Ç–µ —Å–æ–¥–µ—Ä–∂–∏–º–æ–µ: ')))
                    file_res.close()
                    print('–£—Å–ø–µ—Ö')
                except Exception as e:
                    print(e) 
            elif a=='6':
                try:
                    my_file=str(input('–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞: '))
                    my_file+='.txt'
                    file_res=open(my_file,'r')
                    print(file_res.readlines())
                    file_res.close()
                except Exception as e:
                    print(e)
            elif a=='7':
                try:
                    my_file=str(input('–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞: '))
                    my_file+='.txt'
                    os.remove(my_file)
                    print('–§–∞–π–ª —É—Å–ø–µ—à–Ω–æ —É–¥–∞–ª–µ–Ω')
                    print(tree(pathlib.Path.cwd()))
                except Exception as e:
                    print(e)    
            elif a=='8':
                my_file=str(input('–í–≤–µ–¥–∏—Ç–µ –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞: '))
                my_file+='.txt'
                file_end = str(input('–í–µ–¥–∏—Ç–µ –ø—É—Ç—å –∫—É–¥–∞ –∫–æ–ø–∏—Ä—É–µ–º: '))
                shutil.copy(my_file,file_end)
                print('–§–∞–π–ª —É—Å–ø–µ—à–Ω–æ —Å–∫–æ–ø–∏—Ä–æ–≤–∞–Ω')
                print(tree(pathlib.Path.cwd()))
            elif a=='9':
                my_file=str(input('–í–≤–µ–¥–∏—Ç–µ –ø—É—Ç—å —Ñ–∞–π–ª–∞: '))
                my_file+='.txt'
                file_end = str(input('–í–µ–¥–∏—Ç–µ –ø—É—Ç—å –∫—É–¥–∞ –ø–µ—Ä–µ–º–µ—â–∞–µ–º: '))
                shutil.move(my_file,file_end)
                print('–§–∞–π–ª —É—Å–ø–µ—à–Ω–æ –ø–µ—Ä–µ–º–µ—â–µ–Ω')
                print(tree(pathlib.Path.cwd()))
            elif a=="exit":
                byl = False
                print('üò¢Goodbyeüò¢')

            else:
                print('–í—ã–±–µ—Ä–∏—Ç–µ –¥–µ–π—Å—Ç–≤–∏–µ –∏–∑ —Å–ø–∏—Å–∫–∞')
    elif st=='exit':
        byl = False
        print('üò¢Goodbyeüò¢')    
    else:
        print('üò†start - –¥–ª—è –Ω–∞—á–∞–ª–∞ —Ä–∞–±–æ—Ç—ã // exit - –¥–ª—è –≤—ã—Ö–æ–¥–∞üò†')
        creation()        
def main():
    rules()
    tree(pathlib.Path.cwd())
    creation()
    
```
